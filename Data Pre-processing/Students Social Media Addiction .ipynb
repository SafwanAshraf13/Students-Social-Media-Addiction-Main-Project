```python
#Sleep Deficit (8-hours)
df["Sleep_Deficit"] = 8 - df["Sleep_Hours_Per_Night"]
df
```

```python
#Addiction_Category Using Score Range
df["Addiction_Category"] = pd.cut(df["Addicted_Score"],bins = [0,4,8,12,100],labels = ["Low","Moderate","High","Severe"])
df
```

```python
#Sleep vs Social Media Interaction
df["SM_Hours_Per_Sleep_Hour"] = df["Avg_Daily_Usage_Hours"] / df["Sleep_Hours_Per_Night"]
df
```

```python
#SM_Per_SH category using SM_Hours_Per_Sleep_Hour
df["SM_Per_SH Category"] = pd.cut(df["SM_Hours_Per_Sleep_Hour"],bins = [-1,0.75,1.25,2.0,float("inf")],
                                  labels = ["Balanced","Slight Imbalanced","High Imbalanced","Severe Addiction Risk"])
df
```
