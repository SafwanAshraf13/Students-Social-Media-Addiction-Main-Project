```python
#Sleep Deficit (8-hours)
df["Sleep_Deficit"] = 8 - df["Sleep_Hours_Per_Night"]
df
```

```python
#Addiction_Category Using Score Range
df["Addiction_Category"] = pd.cut(df["Addicted_Score"],bins = [0,4,8,12,100],labels = ["Low","Moderate","High","Severe"])
df
```

```python
#Sleep vs Social Media Interaction
df["SM_Hours_Per_Sleep_Hour"] = df["Avg_Daily_Usage_Hours"] / df["Sleep_Hours_Per_Night"]
df
```

```python
#SM_Per_SH category using SM_Hours_Per_Sleep_Hour
df["SM_Per_SH Category"] = pd.cut(df["SM_Hours_Per_Sleep_Hour"],bins = [-1,0.75,1.25,2.0,float("inf")],
                                  labels = ["Balanced","Slight Imbalanced","High Imbalanced","Severe Addiction Risk"])
df
```

Data Visualizations

```python
#Histogram - Daily Usage Hours
plt.figure(figsize = (14,5))
plt.hist(df["Avg_Daily_Usage_Hours"],bins = 10,color = "violet")
plt.title("Distribution Of Daily Social Media Usage Hours")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
#Histogram - Sleep Hours
plt.figure(figsize = (14,5))
plt.hist(df["Sleep_Hours_Per_Night"],bins = 10,color = "violet")
plt.title("Distribution Of Sleep Hours")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
#Bar Chart - Addiction Category
plt.figure(figsize = (14,5))
df["Addiction_Category"].value_counts().plot(kind = 'bar',color = "purple")
plt.title("Addiction Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation = 0)
plt.show()
```

```python
#Bar Chart - Social Media Sleep Category
plt.figure(figsize = (14,5))
df["SM_Per_SH Category"].value_counts().plot(kind = 'bar',color = "purple")
plt.title("Social Media Sleep Hour Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation = 0)
plt.show()
```

```python
#BarPlot  - Mental Health Score by Gender (Platforms)
platforms = ["Instagram","YouTube","WhatsApp"]
df_filtered = df[df['Most_Used_Platform'].isin(platforms)]
plt.figure(figsize=(14,5))
sns.barplot(data = df_filtered,x = "Gender",y = "Mental_Health_Score",hue = "Most_Used_Platform",palette="pastel")
plt.title("Mental Health Score by Gender (Platforms)")
plt.xticks(rotation = 0)
plt.tight_layout()
plt.show()
```

```python
#Scatterplot - Daily Usage vs Sleep Hours
plt.figure(figsize = (14,5))
plt.scatter(df["Avg_Daily_Usage_Hours"],df["Sleep_Hours_Per_Night"],marker = "*",color = "green")
plt.title("Daily Usage Hours vs Sleep Hours")
plt.xlabel(" Usage Score")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
#Scatterplot - Addiction Score vs Sleep Hours
plt.figure(figsize = (14,5))
plt.scatter(df["Addicted_Score"],df["Sleep_Hours_Per_Night"],marker = "*",color = "green")
plt.title("Addiction Score vs Sleep Hours")
plt.xlabel("Addiction Score")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
#PieChart - Addiction Category Distribution
plt.figure(figsize = (7,7))
df["Addiction_Category"].value_counts().plot(kind = "pie",autopct = "%1.1f%%",startangle = 50)
plt.title("Addiction Category Distribution")
plt.ylabel("")
plt.show()
```

```python
#Heatmap - Correlation
plt.figure(figsize = (14,6))
sns.heatmap(df.corr(numeric_only = True),annot = True,cmap = "coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
